* Modify the constructor so that while it still creates the member variable `original`,
  * it creates another member variable called `sanitized`, a list which contains only numbers from the input data.
  * it creates another member variable called `junk`, a list which only contains non-numbers from the input data.
  * the resulting lists must preserve their order

````python
# Your modified code here!
class SaniData:
    def __init__(self, input_data):
        # Verify that input_data is a list
        if not isinstance(input_data, list):
            raise Exception("Must initialize SaniData with a `list` type!")
        
        # Initialize member variables
        self.original = input_data
        self.sanitized = []
        self.junk = []
        
        # Populate sanitized and junk lists
        for item in input_data:
            if isinstance(item, (int, float)):  # Check if the item is a number (int or float)
                self.sanitized.append(item)
            else:
                self.junk.append(item)
````

## Step 3
* Create a function called "max()" which returns the highest value of the list.

Hint: There is a built-in `max` function you can use on lists, but you'll have to check whether or not list is empty!

````python
# Your code here
class SaniData:
    def __init__(self, input_data):
        # Verify that input_data is a list
        if not isinstance(input_data, list):
            raise Exception("Must initialize SaniData with a `list` type!")
        
        # Initialize member variables
        self.original = input_data
        self.sanitized = []
        self.junk = []
        
        # Populate sanitized and junk lists
        for item in input_data:
            if isinstance(item, (int, float)):  # Check if the item is a number (int or float)
                self.sanitized.append(item)
            else:
                self.junk.append(item)
    def max(self):
        # Return the maximum numerical value if sanitized is not empty
        if self.sanitized:
            return max(self.sanitized)
        else:
            return None  # Or you could raise an exception or return a message
````

````python
def test_max_method():
    test_passed = True
    
    # Test case 1: Normal list with numbers
    try:
        data = [1, 2, 3, 4, 5]
        sani_data = SaniData(data)
        assert sani_data.max() == 5, "Max value should be 5."
    except AssertionError as e:
        test_passed = False
        print(f"Test case 1 failed: {e}")

    # Test case 2: List with numbers and other types
    try:
        data = ['x', 1.5, 2, 'y', 99]
        sani_data = SaniData(data)
        assert sani_data.max() == 99, "Max value should be 99."
    except AssertionError as e:
        test_passed = False
        print(f"Test case 2 failed: {e}")

    # Test case 3: Empty list
    try:
        data = []
        sani_data = SaniData(data)
        assert sani_data.max() is None, "Max value should be None for an empty list."
    except AssertionError as e:
        test_passed = False
        print(f"Test case 3 failed: {e}")

    # Test case 4: List with no numbers
    try:
        data = ['a', 'b', 'c']
        sani_data = SaniData(data)
        assert sani_data.max() is None, "Max value should be None for a list with no numbers."
    except AssertionError as e:
        test_passed = False
        print(f"Test case 4 failed: {e}")

    # Test case 5: List with negative and positive numbers
    try:
        data = [-10, 1, 2, 3, -4, 5]
        sani_data = SaniData(data)
        assert sani_data.max() == 5, "Max value should be 5."
    except AssertionError as e:
        test_passed = False
        print(f"Test case 5 failed: {e}")

    # Test case 6: List with floating point numbers
    try:
        data = [1.1, 2.2, 3.3]
        sani_data = SaniData(data)
        assert sani_data.max() == 3.3, "Max value should be 3.3."
    except AssertionError as e:
        test_passed = False
        print(f"Test case 6 failed: {e}")

    # Final results
    if test_passed:
        print("All tests for max method passed!")
    else:
        print("Some tests for max method failed.")

# Run the test function for max method
test_max_method()
````
